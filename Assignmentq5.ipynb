```
month=input("Enter the month:")
winter=["December","January","February"]
spring=["March","April","May"]
summer=["June","July","August"]
autum=["September","October","november"]
if month in winter:
  print("Winter")
if month in spring:
  print("Spring")
elif month in summer:
  print("Summer")
elif month in autum:
  print("Autumn")
```
